```python
import numpy as np
import cv2

def getFrame(video_name, save_path):
    """
    传入视频名称，将图像帧保存到save_path下
    """
    # 读取视频
    video = cv2.VideoCapture(video_name)

    # 获取视频帧率
    fps = video.get(cv2.CAP_PROP_FPS)
    # 获取画面大小
    width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    size = (width, height)

    # 获取帧数
    frame_num = str(video.get(3))
    name = int(math.pow(10, len(frame_num)))
    ret, frame = video.read()
    while ret:
        cv2.imwrite(save_path + str(name) + '.jpg', frame)
        ret, frame = video.read()
        name += 1
    video.release()
    return fps, size
```

```python
import cv2
import glob

path = r"./slow/*.jpg" 
result_name = 'output.mp4'

frame_list = sorted(glob.glob(path))
print("frame count: ",len(frame_list))

fps = 30
shape = cv2.imread(frame_list[0]).shape # delete dimension 3
size = (shape[1], shape[0])
print("frame size: ",size)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(result_name, fourcc, fps, size)

for idx, path in enumerate(frame_list):
    frame = cv2.imread(path)
    # print("\rMaking videos: {}/{}".format(idx+1, len(frame_list)), end = "")
    current_frame = idx+1
    total_frame_count = len(frame_list)
    percentage = int(current_frame*30 / (total_frame_count+1))
    print("\rProcess: [{}{}] {:06d} / {:06d}".format("#"*percentage, "."*(30-1-percentage), current_frame, total_frame_count), end ='')
    out.write(frame)

out.release()
print("Finish making video !!!")
```

```python
import numpy as np
import cv2
import dlib

cap = cv2.VideoCapture('./hw_2.mp4')

fgbg = cv2.createBackgroundSubtractorMOG2(history=1, detectShadows = True)   # MOG2
fgbg1 = cv2.createBackgroundSubtractorMOG2(history=200, detectShadows = False)   # MOG2
bs = cv2.createBackgroundSubtractorKNN(detectShadows=True)
detector = dlib.get_frontal_face_detector()  # Dlib 的人臉偵測器   

fourcc = cv2.VideoWriter_fourcc(*'MJPG') 
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))        # 取得畫面尺寸
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter('./homework2.avi', fourcc, fps, (800, 450))

video = []

while cap.isOpened():
    ret, frame = cap.read()
    if not ret : break
    # print(cap.get(cv2.CAP_PROP_POS_MSEC))
    lo = cap.get(cv2.CAP_PROP_POS_MSEC)
    # print(lo)
    # origin
    if lo < 1000:
        frame1 = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
        cv2.putText(frame1, 'origin', (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)
        video.append(frame1)
        cv2.imshow('hhh', frame1)
        out.write(frame1)
    #MOG
    elif lo >= 1000 and lo < 3000:
        frame1 = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
        fgmask = fgbg.apply(frame1)
        cv2.putText(fgmask, 'MOG', (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv2.LINE_AA)
        image = np.expand_dims(fgmask, axis=2)
        image = np.concatenate((image, image, image), axis=-1)
        video.append(image)
        cv2.imshow('hhh', image)
        out.write(image)
    # MOG
    elif lo >= 3000 and lo < 5000:
        frame1 = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
        fgmask1 = fgbg1.apply(frame1)
        cv2.putText(fgmask1, 'MOG', (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv2.LINE_AA)
        image = np.expand_dims(fgmask1, axis=2)
        image = np.concatenate((image, image, image), axis=-1)
        video.append(image)
        cv2.imshow('hhh', image)
        out.write(image)
    # KNN
    elif lo >= 5000 and lo < 7000:
        frame1 = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
        fg_mask = bs.apply(frame1)
        cv2.putText(fg_mask, 'KNN', (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv2.LINE_AA)
        image = np.expand_dims(fg_mask, axis=2)
        image = np.concatenate((image, image, image), axis=-1)
        video.append(image)
        cv2.imshow('hhh', image)
        out.write(image)
    # RGB
    elif lo >= 7000 and lo < 9000:
        frame1 = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
        frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2RGB)
        frame1 = cv2.flip(frame1, 1)
        cv2.putText(frame1, 'flip+RGB', (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)
        video.append(frame1)
        cv2.imshow('hhh', frame1)
        out.write(frame1)
    elif lo >= 9000 and lo < 11000:
        frame1 = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
        frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
        cv2.putText(frame1, 'Gray', (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)
        video.append(frame1)
        cv2.imshow('hhh', frame1)
        out.write(frame1)       
     
    elif lo >= 11000 and lo < 13000:
        
        frame1 = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
        face_rects, scores, idx = detector.run(frame1, 0)  # 偵測人臉

        for i, d in enumerate(face_rects):                  # 取出所有偵測的結果
            x1 = d.left(); y1 = d.top(); x2 = d.right(); y2 = d.bottom()
            text = f'{scores[i]:.2f}, ({idx[i]:0.0f})'

            cv2.rectangle(frame1, (x1, y1), (x2, y2), (0, 255, 0), 4, cv2.LINE_AA) # 以方框標示偵測的人臉
            cv2.putText(frame1, text, (x1, y1), cv2.FONT_HERSHEY_DUPLEX,    # 標示分數
                0.7, (255, 255, 255), 1, cv2.LINE_AA)
        video.append(frame1)
        cv2.imshow('hhh', frame1)
        out.write(frame1)
    else:    
        break   
            
    k = cv2.waitKey(80) & 0xff
    if k == 27:
        break

# for i in video:
# #     cv2.imshow('original', i)
#     out.write(i)


cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
